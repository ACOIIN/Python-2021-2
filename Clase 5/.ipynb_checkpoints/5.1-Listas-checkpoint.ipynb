# Listas
Las listas en python son una estructura de datos (forma en la cual se guardan los datos) la cual simula una secuencia de elementos. Las listas son dinámicas, es decir que su tamaño puede crecer indefinidamente. 

## 1.1 Crear una lista
Veamos como se crea una lista:

```python
lista = []        # se crea la lista
lista.append(1)   # se añade elemento con .append(x)
lista.append(2)   # se añade elemento con .append(x)
lista
```

También se puede declarar una lista con elementos dentro

```python
lista = [1, 2, 3]
lista
```

Para acceder al elemento en cierta posición se utilizan los corchetes \[  \] al igual que en un string.

```python
lista = [1, 2, 3]
lista[0]
```

Notese que los elementos dentro de la lista no necesariamente deben de ser el mismo tipo de dato entre sí.

```python
lista = ["jojo", True, 3, 3.5555]
lista
```

## 1.2 Métodos y funciones de las listas
Los métodos son funciones propias de los objetos.

```python
# Largo de una lista
len(lista)  
```

```python
# Eliminación de todos los elementos de una lista
lista.clear()
lista
```

```python
# Cuenta cuántas veces hay tal elemento
lista.count(1)
```

```python
# Elimina el primer elemento con igual al valor específico
lista = [1, 2, 3]
lista.remove(1)
lista
```

```python
# Ordena una lista de menor a mayor
lista = [2, 1, 3]
lista.sort()
lista
```

Hay variaos métodos más que pueden encontrar en el siguiente [link](https://www.w3schools.com/python/python_ref_list.asp). 

## 1.3 Recorrer una lista
Para recorrer una lista usaremos lo aprendido en la clase pasada. Mostraremos dos maneras de hacerlo, una con while y otra con for.

### 1.3.1 While

```python
lista = ["jojo", True, 3, 3.5555]
i = 0
while i < len(lista):
    print (lista[i])
    i += 1
```

### 1.3.2 For

```python
lista = ["jojo", True, 3, 3.5555]
for elemento in lista:
    print (elemento)
```

## 2 Listas de listas
Recordando que las listas no prohiben el tipo de dato de los elementos que la componen, se pueden hacer listas compuestas de otras listas. A esto se le conoce como *Matriz*. 

Las matrices tienen largo y ancho como comunmente le llamamos. Más formalmente se dice que una matriz es de N x M, donde N es la cantidad de columnas y M es la cantidad de filas.

### 2.1 Creación de matriz

```python
matriz = [[-2, 5, 6], [5, 2, 7]]
print(matriz)
```

Para que se vea mejor podemos imprimirlo de la siguiente manera:

```python
for fila in matriz:
    print(fila)
```

```python
a = 0          # fila
b = 2          # columna
matriz[a][b]
```

### 2.2 Lista por comprensión
Con fin de hacer la creación de la matriz de forma más automatizada vamos a ingresar el concepto de lista por comprensión.

Las listas por compresión es una forma de definir una lista dada cierta condición.


```python
# Lo que uno normalmente hace
lista = []
for i in range(10):
    lista.append(i)
lista
```

```python
# Su versión en lista por comprensión
lista = [i for i in range(10)]
lista
```

Ahora como requerimos una lista donde sus elementos son listas podemos hacer lo siguiente:

```python
lista = [[1, 1, 2] for i in range(10)]
lista
```

Más generalmente podemos hacer que la lista que damos como argumento esté definida por una lista por comprensión:

```python
fila = [i for i in range(10)]
lista = [fila for i in range(10)]
lista
```

### 3.1 Cómo ingresar una lista con la función input
En python existe una forma fácil de ingresar una lista mediante input y es usando la función split.
La función split() permite transformar un string a una lista, separando por un caracter que se explicita en el argumento (por default es el espacio " ").

Integrando esto con la función input() queda algo así

```python
string = input("Ingrese nombres de paises separados por un espacio\n")
lista = string.split()
print(lista)
```
