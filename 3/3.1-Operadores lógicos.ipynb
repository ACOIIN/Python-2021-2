# Operadores Lógicos
Las operaciones lógicas binaria son `AND` y `OR`. Es decir que se utilizan entre dos elementos. Ej: a + b.

Otra operación que no es binaria es el `not`.

Pasaremos a explicar cada uno.

## 1.`AND` Conjunción
`AND` o `and` que significa en inglés "y", es el operador que se usa cuando se quiere verificar que ambas proposiciones son `True`.

Por ejemplo cuando uno dice: "mañana es sábado y hoy es sábado"

La proposición general es falsa pese a que una de las proposiciones individualmente es verdadera.

### 1.1 Ejemplos

```python
True and True
```

```python
(1 < 3) and ("4" in "hola 4")
```

```python
(1 > 3) and ("4" in "hola 4")
```

## 2.`OR` Disyunción
`OR` o `or` que significa en inglés "o", es el operador que se usa cuando se quiere verificar que alguna de las dos proposiciones es verdad.

Por ejemplo cuando uno dice: "mañana es sábado u hoy es sábado"

La proposición general es verdadera ya que la segunda es verdadera.

### 2.1 Ejemplos

```python
True or False
```

```python
(1 < 3) or ("4" in "hola 4")
```

```python
(1 > 3) or ("4" in "hola 3")
```

## 3. `not` Negación
La negación actúa sobre una proposición lógica y se utiliza para cambiar el valor de verdad de este.

### 3.1 Ejemplos

```python
not True
```

```python
not False
```

```python
not 3 > 2
```

```python
not ("3" in "hola 3")
```

## 4. Uso de `( )`
En python es posible utilizar los parentesis con el propopsito de aislar operaciones lógicas y poder juntarlas con otros más mediante operaciones lógicas.

### 4.1 Ejemplos

```python
False and (True or True)
```

```python
False and True or True
```

```python
(False and True) or True
```

```python
("1" == 1 and 7 > 1) or ((False) and "1" in "1234567")
```
